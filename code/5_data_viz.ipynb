# Data Visualization
In this notebook, I briefly look at the most strongly correlated (positive & negative) words for The West Wing subreddit using horizontal barplots. I essentially use the same code fore each dataset, though I plot more features from the datasets without show-specific words. 

I also engage with wordcloud for the first time.

```python
# Import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Import vectorized datasets from modelling notebook

lr = pd.read_csv('../data/logreg.csv')
lr2 = pd.read_csv('../data/logreg2.csv')
nb = pd.read_csv('../data/multi_nb.csv')
nb2 = pd.read_csv('../data/multi_nb2.csv')
```

```python
# Plots features/probability for logistic regression

plt.figure(figsize=(5, 8))
lr = pd.concat([lr.head(), lr.tail()])
plt.barh(lr['features'], lr['probs'], color='indianred')
plt.yticks(fontsize=15)
plt.xticks(fontsize=8)
plt.xlabel("Probability Coefficient", fontsize=20)
plt.title('5 Strongest +/- Words (LogReg)', fontsize=30);
```

```python
# Plots features/probability for Multinomial Naive Bayes

plt.figure(figsize=(5, 8))
nb = pd.concat([nb.head(), nb.tail()])
plt.barh(nb['features'], nb['probs'], color='dodgerblue')
plt.yticks(fontsize=15)
plt.xticks(fontsize=8)
plt.xlabel("Coefficients", fontsize=20)
plt.title('5 Strongest +/- Words (MultiNB)', fontsize=30);
```

```python
# Plots features/probability for logistic regression

plt.figure(figsize=(5, 10))
lr2 = pd.concat([lr2.head(15), lr2.tail(15)])
plt.barh(lr2['features'], lr2['probs'], color='indianred')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel("Probability Coefficient", fontsize=20)
plt.title('15 Strongest +/- Words (LogReg)', fontsize=30);
```

```python
# Plots features/probability for Multinomial Naive Bayes

plt.figure(figsize=(5, 10))
nb2 = pd.concat([nb2.head(15), nb2.tail(15)])
plt.barh(nb2['features'], nb2['probs'], color='dodgerblue')
plt.yticks(fontsize=15)
plt.xticks(fontsize=5)
plt.xlabel("Coefficients", fontsize=20)
plt.title('15 Strongest +/- Words (MultiNB)', fontsize=30);
```

```python
# Time for some REAL data science! 

# Concatenate my truncated datasets without show-specific words (from above)and put all the text into a string
word_cloud = pd.concat([lr2, nb2])
words = ''
for i in word_cloud['features']:
    words += "".join(i) + " "

# This wordcloud code was adapted from https://www.geeksforgeeks.org/generating-word-cloud-python/
wordcloud = WordCloud(width = 1500, height = 800, 
                background_color ='white',  
                min_font_size = 10).generate(words)
  
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 

plt.show();
```
